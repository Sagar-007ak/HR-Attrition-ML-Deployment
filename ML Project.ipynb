<h3 style="color:seagreen">Data Preprocessing</h3>

```python
# Label encode the target variable
df['Attrition'] = df['Attrition'].map({'No':0,'Yes':1})
```

```python
#Encode Binary Variables
# Encode Binary Variables
df['OverTime'] = df['OverTime'].map({'No': 0, 'Yes': 1})
df['Gender'] = df['Gender'].map({'Male': 0, 'Female': 1})

# Encode Categorical Variables
df['BusinessTravel'] = df['BusinessTravel'].map({
    "Non-Travel": 0,
    "Travel_Frequently": 1,
    "Travel_Rarely": 2
})

df['Department'] = df['Department'].map({
    "Human Resources": 0,
    "Research & Development": 1,
    "Sales": 2
})

df['EducationField'] = df['EducationField'].map({
    "Human Resources": 0,
    "Life Sciences": 1,
    "Marketing": 2,
    "Other": 3,
    "Medical": 4,
    "Technical Degree": 5
})

df['JobRole'] = df['JobRole'].map({
    "Manager": 0,
    "Manufacturing Director": 1,
    "Healthcare Representative": 2,
    "Human Resources": 3,
    "Laboratory Technician": 4,
    "Research Director": 5,
    "Research Scientist": 6,
    "Sales Executive": 7,
    "Sales Representative": 8
})

df['MaritalStatus'] = df['MaritalStatus'].map({
    "Divorced": 0,
    "Married": 1,
    "Single": 2
})


```

<h3 style="color:blueviolet">Targets and Features</h3>

```python
X = df.drop('Attrition', axis=1)
y = df['Attrition']
```

```python
X.shape
```

<h3 style="color:chocolate">Splitting Dataset</h3>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
```

<h3 style="color:red">Resampling Datest to handle imbalance Class using (SMOTE)</h3>

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)

X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
# from sklearn.impute import SimpleImputer
# from imblearn.over_sampling import SMOTE
# from collections import Counter

# # Handle missing values
# imputer = SimpleImputer(strategy='mean')
# X_imputed = imputer.fit_transform(X)

# # SMOTE
# smote = SMOTE(random_state=42)
# X_resampled, y_resampled = smote.fit_resample(X_imputed, y)

# print("Balanced Data:", Counter(y_resampled))
```

```python
from collections import Counter
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
print("Unbalanced Data:",Counter(y))

X_train_smote, y_train_smote = smote.fit_resample(X, y)
print("balanced Data:",Counter(y_train_smote))



```

Important: Only apply SMOTE to the training set, not the test set, to prevent data leakage.

```python
sns.countplot(x=y_train_smote)
plt.title('Class Distribution After SMOTE')
plt.xlabel('Attrition (0=No, 1=Yes)')
plt.ylabel('Count')
plt.show()
```

<h3 style="color:seagreen">Feature Scaling </h3>

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_smote)
X_test_scaled = scaler.transform(X_test)
```

<h3 style="color:royalblue">Logistic Regression</h3>

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_auc_score, roc_curve

model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train_smote)
```

<h3 style="color:royalblue"> Predictions and Probabilities</h3>

```python
y_train_pred = model.predict(X_train_scaled)
y_test_pred = model.predict(X_test_scaled)
y_train_prob = model.predict_proba(X_train_scaled)[:, 1]
y_test_prob = model.predict_proba(X_test_scaled)[:, 1]
```

<h3 style="color:royalblue"> Train Metrics </h3>

```python
print("Train Accuracy:", accuracy_score(y_train_smote, y_train_pred))
print("Train Classification Report:\n", classification_report(y_train_smote, y_train_pred))
print("Train Confusion Matrix:\n", confusion_matrix(y_train_smote, y_train_pred))
print("Train ROC AUC Score:", roc_auc_score(y_train_smote, y_train_prob))
```

<h3 style="color:royalblue"> Test Metrics </h3>

```python
print("Test Accuracy:", accuracy_score(y_test, y_test_pred))
print("Test Classification Report:\n", classification_report(y_test, y_test_pred))
print("Test Confusion Matrix:\n", confusion_matrix(y_test, y_test_pred))
print("Test ROC AUC Score:", roc_auc_score(y_test, y_test_prob))
```

<h3 style="color:royalblue">Confusion Matrix</h3>

```python
cm_train_log = confusion_matrix(y_train_smote, y_train_pred)

cm_test_log = confusion_matrix(y_test, y_test_pred)

plt.figure(figsize=(10, 5))

# Plot confusion matrix for training data
plt.subplot(1, 2, 1)
sns.heatmap(cm_train_log, annot=True, fmt="d", cmap='Blues', xticklabels=['Predicted No', 'Predicted Yes'], yticklabels=['Actual No', 'Actual Yes'])
plt.title('Confusion Matrix - Train')
plt.xlabel('Predicted')
plt.ylabel('Actual')

# Plot confusion matrix for test data
plt.subplot(1, 2, 2)
sns.heatmap(cm_test_log, annot=True, fmt="d", cmap='Blues', xticklabels=['Predicted No', 'Predicted Yes'], yticklabels=['Actual No', 'Actual Yes'])
plt.title('Confusion Matrix - Test')
plt.xlabel('Predicted')
plt.ylabel('Actual')

plt.tight_layout()
plt.show()
```

<h3 style="color:royalblue">ROC Curve</h3>

```python
# Calculate ROC curves and AUC scores
fpr_train, tpr_train, _ = roc_curve(y_train_smote, y_train_prob)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)
auc_train = roc_auc_score(y_train_smote, y_train_prob)
auc_test = roc_auc_score(y_test, y_test_prob)

# Plotting
plt.figure(figsize=(8, 6))
plt.plot(fpr_train, tpr_train, color='blue', label=f'Train ROC (AUC = {auc_train:.2f})')
plt.plot(fpr_test, tpr_test, color='green', label=f'Test ROC (AUC = {auc_test:.2f})')
plt.plot([0, 1], [0, 1], 'k--', label='Random Guess')

# Axis labels and title
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Logistic Regression')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from xgboost import XGBClassifier




# Instantiate models (adjust hyperparameters as needed)
model = LogisticRegression(max_iter=1000)
rf_model = RandomForestClassifier(max_depth=5, n_estimators=100)
svm_model = SVC(probability=True)
best_knn = KNeighborsClassifier()
dt_model = DecisionTreeClassifier(max_depth=5)
xgboost_model = XGBClassifier( max_depth=3,n_estimators=100,learning_rate=0.1,
    subsample=0.8,colsample_bytree=0.8,random_state=42,eval_metric='auc')
ada_boost_model = AdaBoostClassifier()
gnb_model = GaussianNB()

# Train all models
model.fit(X_train_scaled, y_train_smote)
rf_model.fit(X_train_smote, y_train_smote)
svm_model.fit(X_train_scaled, y_train_smote)
best_knn.fit(X_train_scaled, y_train_smote)
dt_model.fit(X_train_smote, y_train_smote)
xgboost_model.fit(X_train_smote, y_train_smote)
ada_boost_model.fit(X_train_scaled, y_train_smote)
gnb_model.fit(X_train_smote, y_train_smote)

# Define models with correct test input type (scaled/unscaled)
models = {
    'Logistic Regression': (model, X_train_scaled, X_test_scaled),
    'Random Forest': (rf_model, X_train_smote, X_test),
    'SVM': (svm_model, X_train_scaled, X_test_scaled),
    'KNN': (best_knn, X_train_scaled, X_test_scaled),
    'Decision Tree': (dt_model, X_train_smote, X_test),
    'XGBoost': (xgboost_model, X_train_smote, X_test),
    'AdaBoost': (ada_boost_model, X_train_scaled, X_test_scaled),
    'GNB': (gnb_model, X_train_smote, X_test)
}

# Initialize results
train_results = []
test_results = []

# Evaluate models
for name, (model, X_train_input, X_test_input) in models.items():
    # ---- Train ----
    y_train_pred = model.predict(X_train_input)
    try:
        y_train_prob = model.predict_proba(X_train_input)[:, 1]
    except:
        y_train_prob = model.decision_function(X_train_input)

    train_results.append({
        'Model': name,
        'Accuracy': accuracy_score(y_train_smote, y_train_pred),
        'ROC AUC': roc_auc_score(y_train_smote, y_train_prob),
        'Precision': precision_score(y_train_smote, y_train_pred, zero_division=0),
        'Recall': recall_score(y_train_smote, y_train_pred, zero_division=0),
        'F1-Score': f1_score(y_train_smote, y_train_pred, zero_division=0)
    })

    # ---- Test ----
    y_test_pred = model.predict(X_test_input)
    try:
        y_test_prob = model.predict_proba(X_test_input)[:, 1]
    except:
        y_test_prob = model.decision_function(X_test_input)

    test_results.append({
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_test_pred),
        'ROC AUC': roc_auc_score(y_test, y_test_prob),
        'Precision': precision_score(y_test, y_test_pred, zero_division=0),
        'Recall': recall_score(y_test, y_test_pred, zero_division=0),
        'F1-Score': f1_score(y_test, y_test_pred, zero_division=0)
    })

# Create DataFrames
train_df = pd.DataFrame(train_results).set_index("Model")
test_df = pd.DataFrame(test_results).set_index("Model")

# Display results
print("====== Train Performance ======")
print(train_df.round(4).sort_values(by='F1-Score', ascending=False))

print("\n====== Test Performance ======")
print(test_df.round(4).sort_values(by='F1-Score', ascending=False))
```

SVM delivers the most consistent and balanced results with high accuracy and F1-score on both train and test sets, making it the most reliable for HR attrition prediction.

GNB (Gaussian Naive Bayes) performs the worst with significantly lower precision, recall, and F1-score, especially on the test set, making it unsuitable for this task.
